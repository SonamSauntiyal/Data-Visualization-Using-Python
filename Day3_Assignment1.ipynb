**Question 1**
Plot using sns.relplot, line plot, box plot on fmri data set


```python
import numpy as np
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
print("Numerical Varibales")
print("Timepoint : ",fmri['timepoint'].unique())
print(fmri.describe())
```

```python
print("Categrical Variable")
print("Subject : ",fmri['subject'].unique())
print("Event   : ",fmri['event'].unique())
print("Region  : ",fmri['region'].unique())
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,hue='event')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,hue='subject')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,hue='region')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue='event')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue='subject')
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,kind='line',hue='region')
```

```python
sns.catplot(x='timepoint',y='signal',data=fmri,kind='box',dodge=False)
sns.catplot(x='timepoint',y='signal',data=fmri,kind='box')
```

```python
sns.catplot(x='timepoint',y='signal',data=fmri,kind='box',hue='region',dodge=False)
sns.catplot(x='timepoint',y='signal',data=fmri,kind='box',hue='region')
```

```python
sns.catplot(x='timepoint',y='signal',data=fmri,kind='box',hue='event',dodge=False)  
sns.catplot(x='timepoint',y='signal',data=fmri,kind='box',hue='region')
```
